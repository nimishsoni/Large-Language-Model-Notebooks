```python
import openai

openai.api_key="Your-API_Key"
```

```python
#Call openAI model

def openai_response(user_message, context, temp = 1):
  newcontext = context.copy()
  newcontext.append({'role':'user', 'content':'question: ' + user_message})

  response = openai.chat.completions.create(model = 'gpt-3.5-turbo', messages = newcontext, temperature = temp)

  return (response.choices[0].message.content)
```

```python
context_user = [{'role':'system', 'content':'You are an expert in tennis'}]

print(openai_response('who won the Wibledon 2019', context_user))
```

```python
def continue_conversation(messages, temperature=0.1):
    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User promnopt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Undate the panels to shjow the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

## Create a Simple Natural Language to SQL translator

```python
def add_prompts_conversation_SQL(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User promnopt.
    context.append({'role':'user', 'content':f"{prompt}."})
    context.append({'role':'system', 'content':f"""YOU ARE A SQL Assistant. \
    If the user ask for something that the result is not an SQL Order \
    just answer something nice and simple asking him, with maximum 10 words, for something that \
    can be solved with SQL."""})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to shjow the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600, styles={'background-color': '#F6F6F6'})))

    return pn.Column(*panels)
```
